### 4. 迴圈與條件判斷:

```for``` loop: 固定次數

````python
fruits = ['apple', 'banana', 'cherry', 'durian'] 

for i in fruits:
    print(i)
````

````python
# 九九乘法表
for i in range(1, 10):
    for j in range(1, 10):
        print(f'{i} * {j} = {i * j}')
    print(" ")
````

```while``` loop: 依照條件停止

````python
fruits = ['apple', 'banana', 'cherry', 'durian'] 

i = 0
while i < 4: # i 可以等於 0, 1, 2, 3
    print(fruits[i])
    i += 1
````

三個搭配語法: ```break``` ,  ```continue```, 與 ```pass```:
- 在迴圈內若出現語法```break```，則:  
    強制跳出<font color='red'> 整個 </font> 迴圈。
- 若出現的是 ```continue```，則:  
   強制跳出<font color='red'> 當前這一個 </font> 迴圈。
- 若出現的是 ```pass```，則:  
    甚麼也不做。

break 圖解

````python
count = 0 # 指標
for char in 'content':
    count += 1
    if char == 't':
        break
    print(char)
    
print("")
print("結束迴圈")
print("共執行 {} 次迴圈。".format(count))
````

continue 圖解

````python
count = 0 # 指標
for char in 'content':
    count += 1
    if char == 't':
        continue
    print(char)
    
print("")
print("結束迴圈")
print("共執行 {} 次迴圈。".format(count))
````

```numpy``` 隨機取值的方法:
- ```np.random.randn()```: 從標準常態分布中取值，括號內放你要取幾個值(以及其維度)
- ```np.random.rand()```: 從隨機樣本```[0, 1)``` 中取值
- ```np.random.randint(a, b, size=(b, c))```: 在 ```[a, b)``` 中取出總共 ```b*c``` 個整數
- ```np.linspace(a, b, c)```: 在 ```a``` 到 ```b``` 之間均勻地撒 ```c``` 個點

````python
a = np.linspace(1, 5, 20)
````
