將圖形加上說明:

````python
plt.plot(months, tpTempH, color='red')
plt.plot(months, tpTempL, color='green')

plt.xlabel("Month")
plt.ylabel("Temperature")
plt.title("The highest and lowest temperature in Taipei")
````

### legend:

提供你數據代表的意思，配合 ```plt.plot9()``` 裡的 ```label``` 參數:

````python
plt.plot(months, tpTempH, color='red', marker='o', label="Highest temperature")
plt.plot(months, tpTempL, color='green', marker='.', label="Lowest temperature")

plt.xlabel("Month")
plt.ylabel("Temperature")
plt.title("The highest and lowest temperature in Taipei")

plt.legend(loc = 'best')
````

````python
fig2 = plt.figure() 
z = np.arange(0, 5, 0.1)
ax0 = fig2.add_subplot(221)
ax0.plot(z, np.sin(z), color='red')

ax1 = fig2.add_subplot(222)
ax1.plot(z, np.cos(z), marker='.')

ax2 = fig2.add_subplot(223)
ax2.plot(z, z**2)

ax3 = fig2.add_subplot(224)
ax3.plot(z, np.log(z))
````

### 練習: 將 COVID 台灣每日新增的確診人數畫出來

````python
request_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
time_series = pd.read_csv(request_url)
print(time_series.shape)

id_cols = time_series.columns[:4]
time_series_long = pd.melt(time_series, id_vars=id_cols, var_name='Date', value_name='Confirmed')
print(time_series_long.shape)

date = pd.to_datetime(time_series_long['Date'])
time_series_long = time_series_long.drop('Date', axis=1)
time_series_long.insert(4, 'Date', date)
time_series_long.head()

tw = time_series_long[time_series_long['Country/Region'] == 'Taiwan*']
daily_increase = np.diff(tw['Confirmed'].values)

date = tw['Date'].values
confirmed = tw['Confirmed'].values
````

藉由上面的 code，我們有下面這些資料:

```date```: 時間軸  
```confirmed```: 每日的總確診人數  
```daily_increase```: 每日新增確診人數
